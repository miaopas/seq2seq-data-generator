```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(6,2,figsize=(20,10))

re = 250
for i in range(6):
    data = np.load(f"Re{re}/p{i+1}.npy").T

    amplitude = data[1]
    t = data[0]
    ax.flatten()[i*2].plot(t, amplitude)


    fourierTransform = np.fft.fft(amplitude)/len(amplitude)
    fourierTransform = fourierTransform[range(int(len(amplitude)/2))]
    tpCount     = len(amplitude)
    values      = np.arange(int(tpCount/2))
    timePeriod = t.max()
    frequencies = values/timePeriod
    
    ax.flatten()[i*2+1].plot(frequencies[:800], fourierTransform[:800])
```
